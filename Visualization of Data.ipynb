## Create a bar chart or histrogram chart to visualize the distrubuation of a categorical or continouse variable 
## i.e THE distrubation of age or gender in a population

```python
# import all dependancy for the Visualization  of data
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import os 
```

```python
# Load the dataset
dataset  = pd.read_csv("titanic_dataset.csv")
```

```python
# head of dataset 
dataset.head(10)
```

```python
# Drop missing values in age column
dataset = dataset.dropna(subset=['Age'])
```

```python
dataset
```

```python
# to visualized the data on bar chart 
```

```python

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Age Distribution (Histogram)
sns.histplot(dataset['Age'], bins=30, kde=True, color='skyblue', ax=axes[0, 0])
axes[0, 0].set_title("Age Distribution (Histogram)")
axes[0, 0].set_xlabel("Age")
axes[0, 0].set_ylabel("Count")

# Gender Distribution (Bar Plot)
sns.countplot(x='Sex', data=dataset, palette='coolwarm', ax=axes[0, 1])
axes[0, 1].set_title("Gender Distribution (Bar Chart)")
axes[0, 1].set_xlabel("Gender")
axes[0, 1].set_ylabel("Count")

# Passenger Class Distribution (Bar Plot)
sns.countplot(x="Pclass", data=dataset, palette='viridis', ax=axes[1, 0])
axes[1, 0].set_title("Passenger Class Distribution (Bar Chart)")
axes[1, 0].set_xlabel("Passenger Class")
axes[1, 0].set_ylabel("Count")

# Survival Rate by Class (Bar Plot)
sns.barplot(x='Pclass', y='Survived', data=dataset, palette='magma', ax=axes[1, 1])
axes[1, 1].set_title("Survival Rate by Passenger Class")
axes[1, 1].set_xlabel("Passenger Class")
axes[1, 1].set_ylabel("Survival Rate")

# Show the plots
plt.tight_layout()
plt.show()
```
